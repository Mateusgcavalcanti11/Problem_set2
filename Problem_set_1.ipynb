```python
# Define the production function using Python's `def` keyword.
# The function computes output using the CES production function,
# and switches to Cobb–Douglas when sigma = 1.

def prod_func(K, L, A, sigma, alpha):
    """
    -  To compute output Y from a CES production function we first need to define the case
    when sigma = 1, the CES function reduces to Cobb–Douglas 
      (can be shown using L'Hôpital's rule). We then after define the CES function
    """
    if sigma == 1:
        return A * (K**alpha) * (L**(1 - alpha))
    else:
        rho = (sigma - 1) / sigma #Exponent in the bracket
        out_ind = sigma / (sigma - 1) #Exponent outside the bracker
        inside_brak = alpha * (K**rho) + (1 - alpha) * (L**rho)
        return A * (inside_brak ** out_ind)
```

Additionally, the curve illustrates that as the values of sigma enlarge we can observe the rate at which production falls for an additional unit of capital decreases. What is meant by this is that for larger values of sigma we can observe more concave curve whilst for the curves where sigma value is larger than 1 we see that the curve continues to rise and are more linear.


##### The reason this occurs lies within the definiton of sigma. Sigma illustrates the substitutability of the inputs (K and L), which means it measures the ease at which a firm can substitute a unit of capital for a unit of labour (or vice-versa) whilst keeping output the same. What this means is that as sigma rises capital is able to substitute for labour in production, and thus capital experiences much diminishing marginal returns at a slower rate. 

##### This is because diminishing marginal returns describes the state where as we increase one input (capital in this case) whilst keeoing another fixed (L = 3) the marginal output of each additional unit of the input will fall. However, the more capital can subsitute for labour the slower the rate of diminishing marginal returns will set in.

### a)

##### In order to find the steady state level of capital and output per worker we first identify that in the steady state (with no population growth), capital stock per worker remians constant. Thus kt+1 = kt = k_ss. As a consequence we rewrite the capital law of motion as: 
##### kt = sAkt^alpha + (1 - delta)kt. 
##### Using this we can rearrange to find that the steady state level of capital per worker is: 
##### k* = (sA/delta)^(1/(1-alpha)

##### To find the equation of ouput per woker we can divide the production function by Lt, and rearrange it to find that the equation for output per worker is Yt/Lt = yt = A(Kt/Lt)^alpha = A(kt)^alpha
##### Firstly, we define the steady state capital per worker function: k* = (sA/delta)^(1/(1-alpha)

```python


def ss_k(s, A, delta, alpha):
    '''Parameters:
    s: Savings rate
    A: Total factor peoductivity 
    alpha: Capital share of income 
    delta: Depreciation

    Returns:
    k_ss : Steady state capital per worker
    y_ss : Steady state output per worker
    '''
    inside_bracket = (s*A)/delta
    expo = 1/(1 - alpha)
    return (inside_bracket)**expo

k_ss = ss_k(0.35, 4, 0.07, 0.3) # Plug in the parameters provided

print("The steady state capital per worker (k) is:", k_ss) 

def ss_y(A, k, alpha):
    return A*(k**alpha)

# Below we write the code that solves the steady state ouput per worker function

y_ss = ss_y(4, k_ss, 0.3)

print("The steady state output per worker (y) is:", y_ss)
```

```python

def ss_k(s, A, delta, alpha):
    inside_bracket = (s*A)/delta
    expo = 1/(1 - alpha)
    return (inside_bracket)**expo
    
alpha = 0.3
delta = 0.07
A = 4.0                  
s = 0.35
A_new = 0.75 * A          
eps = 0.5 

k_ss = ss_k(0.35, 4, 0.07, 0.3)
k_ss_new = ss_k(s, A_new, delta, alpha) 

print("The old steady state capital per worker is:", k_ss) 
print("The new steady state capital per worker is:", k_ss_new) 

def ss_y(A, k, alpha):
    return A*(k**alpha)
    
y_ss = ss_y(4, k_ss, 0.3)
y_ss_new = ss_y(A_new, k_ss_new, alpha)

print("The old steady state output per worker is:", y_ss)
print("The new steady state output per worker is:", y_ss_new)


k_path = [k_ss, k_ss, k_ss]   
y_path = [y_ss, y_ss, y_ss]   


k_current = k_path[-1]
t = 2
max_steps = 10000  

while True:
  
    k_next = s * A_new * (k_current ** alpha) + (1 - delta) * k_current
    y_next = ss_y(A_new, k_next, alpha)

    k_path.append(k_next)
    y_path.append(y_next)

    t += 1
    if abs(k_next - k_ss_new) < eps or len(k_path) > max_steps:
        break

    k_current = k_next

print(f"Converged in = {len(k_path)-1} with |k_t - k*_new| = {abs(k_path[-1]-k_ss_new):.3f} < {eps}")


periods = np.arange(len(k_path))


fig, axes = plt.subplots(1, 2, figsize=(10, 4), sharex=True)    

# Capital per worker
axes[0].plot(periods, k_path, marker="o", label="k")
axes[0].set_title("Path to new steady state of capital per worker")
axes[0].axhline(y=k_ss,     linestyle="--", alpha=0.6, label="old k steady state")
axes[0].axhline(y=k_ss_new, linestyle="--", alpha=0.6, label="new k steady state")
axes[0].set_ylabel("Capital per worker")
axes[0].legend()
axes[0].grid(True)

# Output per worker
axes[1].plot(periods, y_path, marker="o", label="y")
axes[1].set_title("Path to new steady state of output per worker")
axes[1].axhline(y=y_ss,     linestyle="--", alpha=0.6, label="old y steady state")
axes[1].axhline(y=y_ss_new, linestyle="--", alpha=0.6, label="new y steady state")
axes[1].set_xlabel("Period t")
axes[1].set_ylabel("Output per worker")
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

### c)

```python
# Solow model transition with temporary savings shock
# Cobb–Douglas: y_t = A * k_t^alpha
# Law of motion: k_{t+1} = s_t * A * k_t^alpha + (1 - delta) * k_t


alpha = 0.3
delta = 0.07
A = 4.0
rho = (1.0/(1.0 - alpha))
s_baseline = 0.35
s_shock = 0.60          
t_shock_start = 3       
t_shock_end = 10

T = 100                 


def ss_k(s, A, delta, alpha):
    return ((s * A) / delta) ** rho

def y_of(A, k, alpha):
    return A * (k ** alpha)

#Initial condition
k0 = ss_k(s_baseline, A, delta, alpha)

# Path
k_path = [0.0] * (T + 1)
y_path = [0.0] * (T + 1)

k_path[0] = k0
y_path[0] = y_of(A, k_path[0], alpha)

for t in range(T): 
    s_t = s_shock if (t_shock_start <= t <= t_shock_end) else s_baseline
    k_next = s_t * A * (k_path[t] ** alpha) + (1.0 - delta) * k_path[t]
    k_path[t + 1] = k_next
    y_path[t + 1] = y_of(A, k_next, alpha)


t_grid = np.arange(T + 1)

plt.figure(figsize=(8, 4.5))
plt.plot(t_grid, y_path, marker="o", markersize=3, label="y")
y_ss_baseline = y_of(A, ss_k(s_baseline, A, delta, alpha), alpha)
plt.axhline(y_ss_baseline, linestyle="--", alpha=0.6, label="baseline output y steady state")

plt.title("Output per Worker, showing the path after positive savings shock")
plt.xlabel("Period t")
plt.ylabel("Output per worker")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
